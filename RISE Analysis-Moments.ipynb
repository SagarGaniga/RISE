```python
import numpy as np
import numpy.random as random
import pandas as pd
import pylab 
import scipy.stats as stats
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
from datetime import datetime
from dateutil.parser import parse
```

```python
df = pd.read_excel("Data/Moments Report.xls")
# df.head()
```

```python
f = "%Y-%m-%d %H:%M:%S"
for i, obj in df.iterrows():
    date1 =  datetime.strptime(obj[-1], f)
    date2 = datetime.strftime(date1, '%Y-%m-%d')
    df.set_value(i, "Date", date2)
    df.set_value(i, "index", i)
current = df
current.head()
```

```python
data = df["Date"].value_counts()
data.plot()

plt.show()
```

```python
top = df["By Name"].value_counts()

top1 = top.head(5)

# data["By Name"].values_count()
top1.plot()

plt.show()
```

```python
for obj in top1.iteritems():
    person = df.loc[df['By Name'] == obj[0]]
    freq_date = person["Date"].value_counts().index.tolist()
    freq_count = person["Date"].value_counts().values.tolist()
#     fr = {'Date': freq_date, 'Count': freq_count}
#     df_fr = pd.DataFrame(data=fr)
#     sorted_df = df_fr.sort_values(by=['Date'])
#     print(sorted_df)
#     sorted_df.plot()
#     dates = matplotlib.dates.date2num(freq_date)
    plt.plot_date(freq_date, freq_count)
    plt.xlabel(obj[0])
    plt.show()
```
